## 2. Explore the Data Structure

Let's examine the first few rows, data types, and summary statistics of our product data.

```python
# Display first 5 rows
print("First 5 products:")
display(df.head())

print("\nData types:")
print(df.dtypes)

print("\nSummary statistics:")
display(df.describe())

print("\nUnique categories:")
print(df['category'].unique())

print(f"\nTotal products: {len(df)}")
print(f"Categories: {df['category'].nunique()}")
```

## 3. Filter Products by Category

Let's filter the DataFrame to show products from specific categories like 'electronics' and 'men's clothing'.

```python
# Filter electronics products
electronics = df[df['category'] == 'electronics']
print("Electronics products:")
display(electronics[['id', 'title', 'price', 'rating_rate']])

# Filter men's clothing
mens_clothing = df[df['category'] == "men's clothing"]
print("\nMen's clothing products:")
display(mens_clothing[['id', 'title', 'price', 'rating_rate']])

# Filter products with rating above 4.0
high_rated = df[df['rating_rate'] > 4.0]
print(f"\nProducts with rating > 4.0 ({len(high_rated)} products):")
display(high_rated[['title', 'category', 'price', 'rating_rate']])
```

## 4. Calculate Average Prices by Category

Group the data by category and compute the average price for each group.

```python
# Calculate average price by category
avg_price_by_category = df.groupby('category')['price'].agg(['mean', 'min', 'max', 'count'])
avg_price_by_category = avg_price_by_category.round(2)
print("Average prices by category:")
display(avg_price_by_category)

# Sort by average price
avg_price_by_category_sorted = avg_price_by_category.sort_values('mean', ascending=False)
print("\nCategories sorted by average price:")
display(avg_price_by_category_sorted)

# Most expensive product in each category
most_expensive = df.loc[df.groupby('category')['price'].idxmax()]
print("\nMost expensive product in each category:")
display(most_expensive[['category', 'title', 'price']])
```

## 5. Analyze Product Ratings

Extract and analyze the rating data, including calculating mean ratings and identifying products with the highest or lowest ratings.

```python
# Overall rating statistics
print("Rating statistics:")
print(f"Average rating: {df['rating_rate'].mean():.2f}")
print(f"Median rating: {df['rating_rate'].median():.2f}")
print(f"Highest rating: {df['rating_rate'].max():.2f}")
print(f"Lowest rating: {df['rating_rate'].min():.2f}")

# Average rating by category
avg_rating_by_category = df.groupby('category')['rating_rate'].agg(['mean', 'count']).round(2)
print("\nAverage rating by category:")
display(avg_rating_by_category)

# Top 5 highest rated products
top_rated = df.nlargest(5, 'rating_rate')
print("\nTop 5 highest rated products:")
display(top_rated[['title', 'category', 'rating_rate', 'rating_count']])

# Bottom 5 lowest rated products
bottom_rated = df.nsmallest(5, 'rating_rate')
print("\nBottom 5 lowest rated products:")
display(bottom_rated[['title', 'category', 'rating_rate', 'rating_count']])

# Products with most reviews
most_reviewed = df.nlargest(5, 'rating_count')
print("\nProducts with most reviews:")
display(most_reviewed[['title', 'category', 'rating_rate', 'rating_count']])
```

## 6. Visualize Price Distribution

Create visualizations to show the distribution of prices across different categories using matplotlib and seaborn.

```python
# Set up the plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Fake Store API Product Analysis', fontsize=16, fontweight='bold')

# 1. Price distribution histogram
axes[0, 0].hist(df['price'], bins=20, edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Price Distribution')
axes[0, 0].set_xlabel('Price ($)')
axes[0, 0].set_ylabel('Number of Products')
axes[0, 0].axvline(df['price'].mean(), color='red', linestyle='--', label=f'Mean: ${df["price"].mean():.2f}')
axes[0, 0].legend()

# 2. Box plot of prices by category
sns.boxplot(x='category', y='price', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Price Distribution by Category')
axes[0, 1].set_xlabel('Category')
axes[0, 1].set_ylabel('Price ($)')
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Rating distribution
axes[1, 0].hist(df['rating_rate'], bins=10, edgecolor='black', alpha=0.7, color='green')
axes[1, 0].set_title('Rating Distribution')
axes[1, 0].set_xlabel('Rating')
axes[1, 0].set_ylabel('Number of Products')
axes[1, 0].axvline(df['rating_rate'].mean(), color='red', linestyle='--', label=f'Mean: {df["rating_rate"].mean():.2f}')
axes[1, 0].legend()

# 4. Scatter plot: Price vs Rating
scatter = axes[1, 1].scatter(df['rating_rate'], df['price'], alpha=0.6, c=df['rating_count'], cmap='viridis')
axes[1, 1].set_title('Price vs Rating (Color = Review Count)')
axes[1, 1].set_xlabel('Rating')
axes[1, 1].set_ylabel('Price ($)')
plt.colorbar(scatter, ax=axes[1, 1], label='Review Count')

plt.tight_layout()
plt.show()

# Additional visualization: Category distribution
plt.figure(figsize=(10, 6))
category_counts = df['category'].value_counts()
plt.bar(category_counts.index, category_counts.values, color='skyblue', edgecolor='black')
plt.title('Number of Products by Category')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.show()
```
